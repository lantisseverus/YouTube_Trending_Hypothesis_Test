## Correlation among variables

I will create a heatmap to present the correlation among views, likes, dislikes, and comment_count.

First, I will turn likes, dislikes, and comment_count into ratios based on views.

````python
yt_us_jp_merge['like_rate'] =  yt_us_jp_merge ['likes'] / yt_us_jp_merge['views'] * 100
yt_us_jp_merge['dislike_rate'] =  yt_us_jp_merge ['dislikes'] / yt_us_jp_merge['views'] * 100
yt_us_jp_merge['comment_rate'] =  yt_us_jp_merge ['comment_count'] / yt_us_jp_merge['views'] * 100
````

````python
sns.heatmap(yt_us_jp_merge[['like_rate', 'dislike_rate', 'comment_rate', "category_name"]].corr(), annot=True)
plt.title("Correlation Matrix among Like Rate, Dislike Rate, and Comment Count Rate based on Category")
plt.show()
````

From the correlation matrix, we can tell that likes and the number of comments is of moderate association, which implies that favoring the video may encourage the audience to leave a comment. By contrast, the correlation between dislikes and comments is of mild association, which implies that antipathy toward the video may still stimulate the audience to leave comments but the incentive may not be as strong as favoring one.

## Hypothesis Testing: Which country is more inclined to leave comment?

I would like to perform a simple hypothesis test to understand if the US audience is more inclined to leave comments than the audience in Japan.

**Null Hypotheis:** The comment rate is not significantly different in the US and Japan.

**Alternative Hypothesis:** The comment rate in US is greater than the comment rate in Japan.

**Alpha Level = 0.05**

````python
yt_jp_sub["country"] = "JP"
yt_us_sub["country"] = "US"
yt_jp_sub["comment_rate"] = yt_jp_sub["comment_count"] / yt_jp_sub["views"] * 100
yt_us_sub["comment_rate"] = yt_us_sub["comment_count"] / yt_us_sub["views"] *100
````

````python
%pip install scipy
from scipy.stats import mannwhitneyu
````

````python
scipy_result = mannwhitneyu(x = yt_us_sub["comment_rate"],
                            y = yt_jp_sub["comment_rate"],
                            alternative= "greater")
````

````python
if scipy_result.pvalue <= 0.05:
    result = "reject"
else:
    result = "fail to reject"

result_dict = {"p_val": scipy_result.pvalue, "result": result}

print(result_dict)
````

From the above statistics result we know that the p-value is lower than alpha 0.05, which suggests we have significant evidence to reject the null hypothesis and say that the comment rate in the US is higher than the comment rate in Japan.
